4. Задание (в программе)
Решите систему уравнений:

```python
# Решите систему уравнений:y = x2 – 1    exp(x) + x∙(1 – y) = 1
x = np.linspace(-np.pi, np.pi, 201)
y = np.linspace(-np.pi, np.pi, 201)
f, ax = plt.subplots(figsize=(8, 8))

yx1 = x ** 2 - y -1
yx2 = np.exp(x) + x * (1 - y) - 1
plt.plot(x, yx1);
plt.plot(x, yx2);
ax.vlines(0, -15, 15, color='r', linestyle  = '--');
ax.vlines(-0.5, -15, 15, color='r', linestyle  = '--');
# на этом графике видно, что решение системы уравнений находиться в интервале от -0.5 до 0
```

```python
from sympy import Symbol, nsolve, exp
x = Symbol('x')
y = Symbol('y')
yx1 = x ** 2 - y -1
yx2 = exp(x) + x * (1 - y) - 1

nsolve((yx1, yx2), (x, y), (-0.5, 0))
```

```python
# Решите систему уравнений: y = x2 – 1     exp(x) + x∙(1 – y) > 1

x = np.linspace(-np.pi, np.pi, 201)
y = np.linspace(-np.pi, np.pi, 201)
f, ax = plt.subplots(figsize=(8, 8))

yx1 = x ** 2 - y -1
yx2 = np.exp(x) + x * (1 - y) - 1
plt.plot(x, yx1);
plt.plot(x, yx2);
ax.vlines(0, -15, 15, color='r', linestyle  = '--');
ax.vlines(1, -15, 15, color='r', linestyle  = '--');
# на этом графике видно, что решение системы уравнений находиться в интервале от -0.5 до 0
```
